<h1>Simple Linear Regression</h1>

<h3>This notebook focuses on the creation of a Simple Linear Regression Model</h3>

<h3>Preprocessing as described in Part 1 of this course</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

dataFilePath = '../../data/Part 2 - Regression/Section 4 - Simple Linear Regression/Salary_Data.csv'
dataset = pd.read_csv(dataFilePath)
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, -1].values

xTrain, xTest, yTrain, yTest = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

<h3>Fitting Simple Linear Regression to the Training set</h3>

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(xTrain, yTrain)
```

<h3>Predicting the Test set Results</h3>

```python
yPred = regressor.predict(xTest)
```

<h3>Visualizing Results</h3>

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)
plt.scatter(xTrain, yTrain, color = 'red', label = 'Observed')
plt.plot(xTrain, regressor.predict(xTrain), color = 'blue', label = 'Fit')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (8,6)
plt.scatter(xTest, yTest, color = 'red', label = 'Observed')
plt.plot(xTrain, regressor.predict(xTrain), color = 'blue', label = 'Fit')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
```
